### Задание 1

```python
import json

purchases = {}
with open('purchase_log.txt', 'r', encoding='utf-8') as file:
    file.readline()
    for line in file:
        obj = json.loads(line)
        user_id = obj['user_id']
        category = obj['category']
        purchases[user_id] = category

i = 0
res = list(purchases.values())
print(res[0])
while i != 2:
    print(res[i])
    i += 1
```
